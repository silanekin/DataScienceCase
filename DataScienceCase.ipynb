```python
ct = pd.crosstab(df["Cinsiyet"],df["Bolum"])
sns.heatmap(ct, annot=True, cmap="Blues")
plt.title("Cinsiyet-Bolum Frekansları")
plt.show()
```

```python
bins = [2, 18, 37, 56, 75, 92]
labels = ["2-18","19-37", "38-56","57-75", "76-92"]
df["YasAralik"] = pd.cut(df["Yas"], bins=bins, labels=labels, right=False)
```

```python
ct = pd.crosstab(df["TedaviSuresi"],df["YasAralik"])
sns.heatmap(ct, annot=True,fmt=".1f", cmap="Blues")

plt.title("TedaviSuresi-Yas Aralik Araliklari Frekansları")
plt.show()
```

```python
df["KanGrubu"] = df["KanGrubu"].fillna("Bilinmiyor")
df["KronikHastalik"] = df["KronikHastalik"].fillna("Yok")

# Kronik hastalıkların top 15'ini al
top_kronik = df["KronikHastalik"].value_counts().nlargest(30).index
df["KronikTop"] = df["KronikHastalik"].where(df["KronikHastalik"].isin(top_kronik), "Diğer")

ct = pd.crosstab(df["KanGrubu"], df["KronikTop"])

# Heatmap
plt.figure(figsize=(15,6))
sns.heatmap(ct, annot=True, fmt="d", cmap="Oranges")
plt.title("Kan Grubu - Kronik Hastalık (Top 30) Frekansı")
plt.ylabel("Kan Grubu")
plt.xlabel("Kronik Hastalık")
plt.show()
```

```python
df["KronikVarMi"] = df["KronikHastalik"].apply(lambda x: "Var" if x not in [None, "", "Yok"] else "Yok")
df["KronikVarMi"] = df["KronikVarMi"].fillna("Yok")
```

```python
ct = pd.crosstab([df["YasAralik"], df["Cinsiyet"]], df["KronikVarMi"])

plt.figure(figsize=(8,6))
sns.heatmap(ct, annot=True, fmt="d", cmap="Purples")
plt.title("Yaş Aralığı + Cinsiyet vs Kronik Hastalık Durumu")
plt.ylabel("Yaş Aralığı - Cinsiyet")
plt.xlabel("Kronik Hastalık Var/Yok")
plt.show()
```

```python
plt.figure(figsize=(16,8))
sns.countplot(x='UygulamaSuresi', data=df)
```

```python
df["UygulamaYerleri"] = df["UygulamaYerleri"].fillna("Bilinmiyor")
top_uygulamayeri = df["UygulamaYerleri"].value_counts().nlargest(10).index
df["Uygulamayertop"] = df["UygulamaYerleri"].where(df["UygulamaYerleri"].isin(top_uygulamayeri), "Diğer")
plt.figure(figsize=(20,8))
sns.countplot(x='Uygulamayertop', data=df)
```

```python
ct = pd.crosstab(df["Uygulamayertop"], df["KronikVarMi"])

plt.figure(figsize=(15,6))
sns.heatmap(ct, annot=True, fmt="d", cmap="Oranges")
plt.title("Uygulama yerleri top(10) - Kronik Hastalık (Top 30) Frekansı")
plt.ylabel("Uygulama Yerleri ")
plt.xlabel("Kronik Hastalık Var/Yok")
plt.show()
```
